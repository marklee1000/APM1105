```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

#Load the dataset
df = pd.read_csv("Spotify_Youtube.csv")

#Display the first 10 rows of the data
df.head(10)
```

```python
#Check and display the column names.
df.columns
```

```python
#Display the info of the dataset and make description of the column counts
df.info()
```

```python
# Fill NA in likes with 0, NA in comments with 0
df.fillna({"Likes": 0, "Comments": 0}, inplace=True)

#and drop others.
df.dropna(inplace=True)
```

```python
df.describe()
```

```python
#Explore album type and display counts for album, single, and compilation.
album_type_counts = df["Album_type"].value_counts()
print(album_type_counts)
```

```python
#Create a pie chart for Album Types as a Title and make some remarks.
labels = album_type_counts.index.tolist()
sizes = album_type_counts.values.tolist()

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("Album Types")
plt.axis('equal')
plt.show()
```

The pie chart above shows that 72.3% in the dataset are albums, 24.1% are singles, and 3.6% are compilations.

```python
#Analyze songs by artist and views-youtube, stream-spotify.
```

```python
#Create bar charts for Top 10 Artists on Youtube and Top 10 Artists on Spotify in 1 row of display.

# Get top 10 artists by views on YouTube
top_artists_yt = df.groupby('Artist')['Views'].sum().nlargest(10)

# Get top 10 artists by streams on Spotify
top_artists_sp = df.groupby('Artist')['Stream'].sum().nlargest(10)

# Create subplots
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

# Bar chart for top 10 artists on YouTube
axs[0].barh(top_artists_yt.index, top_artists_yt.values, color='red')
axs[0].set_title('Top 10 Artists on YouTube')
axs[0].set_xlabel('Views')
axs[0].invert_yaxis()

# Bar chart for top 10 artists on Spotify
axs[1].barh(top_artists_sp.index, top_artists_sp.values, color='green')
axs[1].set_title('Top 10 Artists on Spotify')
axs[1].set_xlabel('Streams')
axs[1].invert_yaxis()

# Display the plots
plt.tight_layout()
plt.show()
```

Make some remarks about the Top 10 Artists on Youtube and Top 10 Artists on Spotify.

```python
# Analyze Top 10 songs on the basis of views, comments, and likes: with titles Top 10 Songs with Most Views, Top 10 Songs with Most Comments, and Top 10 Songs with Most Likes.

# Top 10 songs with most views
top_views = df[['Track', 'Views']].sort_values('Views', ascending=False).head(12)

# Create bar chart
plt.figure(figsize=(12,6))
plt.bar(top_views['Track'], top_views['Views'])
plt.title('Top 10 Songs with Most Views')
plt.xlabel('Song Name')
plt.ylabel('Number of Views (in millions)')
plt.xticks(rotation=90)
plt.show()

# Top 10 songs with most comments
top_comments = df[['Track', 'Comments']].sort_values('Comments', ascending=False).head(12)

# Create bar chart
plt.figure(figsize=(12,6))
plt.bar(top_comments['Track'], top_comments['Comments'])
plt.title('Top 10 Songs with Most Comments')
plt.xlabel('Song Name')
plt.ylabel('Number of Comments (in thousands)')
plt.xticks(rotation=90)
plt.show()

# Top 10 songs with most likes
top_likes = df[['Track', 'Likes']].sort_values('Likes', ascending=False).head(14)

# Create bar chart
plt.figure(figsize=(14,6))
plt.bar(top_likes['Track'], top_likes['Likes'])
plt.title('Top 10 Songs with Most Likes')
plt.xlabel('Song Name')
plt.ylabel('Number of Likes (in millions)')
plt.xticks(rotation=90)
plt.show()
```

Make some remarks about the Top 10 Songs with Most Views, Top 10 Songs with Most Comments, and Top 10 Songs with Most Likes.

```python
# Display a bar chart of the Top 10 Songs by Stream with their Energy, Danceability, and Acousticness and make some remarks about it.

# Get the top 10 songs by stream
top10_streams = df.nlargest(14, 'Stream')

# Melt the dataframe to long format
top10_melted = pd.melt(top10_streams, id_vars=['Track'], value_vars=['Energy', 'Danceability', 'Acousticness'])

# Create bar chart
sns.set_style('whitegrid')
sns.set_palette('husl')
plt.figure(figsize=(14, 6))
ax = sns.barplot(x='Track', y='value', hue='variable', data=top10_melted)
ax.set(xlabel='Song', ylabel='Value', title='Top 10 Songs by Stream with Energy, Danceability, and Acousticness')
plt.xticks(rotation=90)
plt.legend(title='Variable', loc='upper right')
plt.show()
```

```python
# Determine the top 10 artists on the Spotify platform whose songs have more likes and display a bar chart with a title Number of Likes per Artist and make some remarks.

# Top 10 Artists with Most Likes
top10_likes = df.groupby('Artist')['Likes'].sum().nlargest(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top10_likes.index, y=top10_likes.values)
plt.xticks(rotation=90)
plt.xlabel('Artist')
plt.ylabel('Number of Likes')
plt.title('Number of Likes per Artist')
plt.show()
```

```python
# Explore single releases by artists and identify top performers based on likes and display a table for each artist with their Views and Likes as the other columns.

# Group data by artists and filter for singles only
singles = df[df['Album_type'] == 'single'].groupby('Artist').sum()

# Sort by likes and get top performers
top_performers = singles.sort_values('Likes', ascending=False)[:10]

# Display table with views and likes for each artist
top_performers[['Views', 'Likes']]
```

```python
# Exploring compilation releases by artists and identify top performers based on views and likes and display a table for each artist with their Views and Likes as the other columns. Make some remark about the table.

# Filter data for compilation releases and group by artist
compilations = df[df['Album_type'] == 'compilation'].groupby('Artist')

# Calculate total views and likes for each artist
compilation_views = compilations['Views'].sum()
compilation_likes = compilations['Likes'].sum()

# Combine views and likes into a single table
compilation_table = pd.concat([compilation_views, compilation_likes], axis=1)
compilation_table.columns = ['Views', 'Likes']

# Sort table by views and get top performers
top_performers_comp = compilation_table.sort_values('Views', ascending=False)[:10]

# Display table with views and likes for each artist
top_performers_comp
```

```python
# Identify the top 10 channels based on views and display a bar chart of it and make some remarks.

# Top 10 Channels with Most Views
top10_likes = df.groupby('Channel')['Views'].sum().nlargest(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top10_likes.index, y=top10_likes.values)
plt.xticks(rotation=90)
plt.xlabel('Channel')
plt.ylabel('Views')
plt.title('Top 10 Channels with Most Views')
plt.show()
```

```python
'''
1. Write a Python function to multiply all the numbers in a list.

Sample List : (5, 6, -1, 2, 4, -2, -10, 7)
'''
import numpy

""" Example """

samplelist = [5, 6, -1, 2, 4, -2, -10, 7]
sampleproduct = numpy.prod(samplelist)
print("Sample List: ", samplelist,)
print("Sample Product: ", sampleproduct,"\n")

""" Function Proper (w/ user input)"""

def list_product():
    true_list = []
    y = 0
    while y<8:
        list_element = int(input('Enter a number: '))
        true_list.append(list_element)
        y = y + 1
    print("List is complete!\n")
    true_product = numpy.prod(true_list)
    print("Entered List: ", true_list,)
    print("Computed Product: ", true_product,)

list_product()
```

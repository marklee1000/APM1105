1. words.txt (A - F)

```python
import requests

# URL of the raw dataset file on GitHub
url = 'https://raw.githubusercontent.com/AllenDowney/ThinkPython2/master/code/words.txt'

# Send a GET request to the URL
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Save the content of the response to a file
    with open('words.txt', 'wb') as f:
        f.write(response.content)
        print('Dataset saved to file.')
else:
    print('Failed to download dataset.')

# Opening the file
with open(r'words.txt','r') as file:
    data = file.read()
    lines = data.split()

# A. Print the total number of words in the file.
number_of_words = 0
number_of_words += len(lines)
print("\nTotal Number of Words: ",number_of_words)

# B. Print the longest word in the file.
longestWordLength = len(max(lines, key=len))
result_longest = [textword for textword in lines if len(textword) == longestWordLength]

print("\nThe following are the longest words from a text file:")
print(result_longest)
print("\n")

# C. Print the number of words that start with a certain letter (use user input).
cword_count = 0
cLetter = input('Enter a letter: ')

for data in lines:
    if data[0]==cLetter:
        cword_count += 1

print(cword_count, "words start with the letter",cLetter,)
print("\n")

# D. Print all the words that contain a certain substring (use user input).
d_sub = input('Enter a substring: ')
print(f"Words that contain '{d_sub}' are the ff:")
for data in lines:
    if d_sub in data:
        print (data)

# E. Print all the words that are palindromes (i.e., read the same backwards and forwards).
def checkPalin(word):
    if word.lower() == word.lower()[::-1]:
        return True
    
def countPalin(str):
    count = 0

    for data in lines:
        if (checkPalin(data)):
            count += 1
            print (data)

print("\nWords that are palindromes:")
countPalin(lines)

# F. Count the frequency of each letter.
freqs = {}
with open(r'words.txt','r') as f:
    for line in f:
        for char in line:
            if char.isalpha():
                if char in freqs:
                    freqs[char] += 1
                else:
                    freqs[char] = 1

print("\nFrequency of each letter:")
print(freqs)
```

2. Write a program on the Fibonacci sequence using both iterations and recursion, and including rendering time to compare the performance of both approaches.

```python
import time

def fibo_iter(n):
    n1, n2 = 1, 1
    count = 2

    if n <= 0:
        print("Please enter a positive integer")
    elif n == 1 or n == 2:
        print("\nIterative approach result:")
        print(n1)
    else:
        print("\nIterative approach result: ")
        while count < n:
            nth = n1 + n2
            n1 = n2
            n2 = nth
            count += 1
        print(nth)

def fibo_recur(n):
    if n <= 0:
        print("Please enter a positive integer")
    elif n == 1 or n == 2:
        return 1
    else:
        return(fibo_recur(n-1) + fibo_recur(n-2))

n = int(input('n = '))
start = time.time()
fibo_iter(n)
end = time.time()
print(f"Time taken (iterative approach): {end-start} seconds\n")

print("Recursive approach result: ")
start = time.time()
print(fibo_recur(n))
end = time.time()
print(f"Time taken (recursive approach): {end-start} seconds\n")
```

3. Write a program to find the solution to the equation x^3 + 5x = 20 using iterative formula x_(n+1) = cbrt(20 - 5x_n) with initial value x_0 = 2, giving the answer to 3 decimal places. (Note: Your program should print the value for x_0, x_1, ... and stops at the x_n when x_n â‰… x_(n-1) at the 3 decimal places.

```python
def eqn(x_zero):
    n_count = 0
    x_n = 0
    stopper = x_zero
    while x_n != stopper:
        stopper = round(x_zero,3)
        print(f"x_{n_count} = {stopper}")
        x_n = (20-5*x_zero)**(1/3)
        x_zero = x_n
        x_n = round(x_n,3)
        n_count += 1
    x_zero = round(x_zero,3)
    print(f"x_{n_count} = {x_zero}")
    print(f"Therefore, a solution to x^3 + 5x = 20 is x = {x_zero}")
eqn(2) #initial value x_0 = 2
```
